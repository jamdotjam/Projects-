## Functions

```python
### FUNCTIONS ###
# Header
def header(msg):
    print('-' * 50)
    print('[' + msg + ']')


# Expected Wins
def pythagoras(dataframe, PS, PA):
    PF = dataframe[PS].sum()
    PA = dataframe[PA].sum()
    GP = len(dataframe[PS])
    EXP = 1.5 * math.log((PF + PA) / GP)
    P = (PF ** EXP) / (PF ** EXP + PA ** EXP) * GP
    P = round(P, 3)
    P = str(P)
    return print('EXPECTED WINS: ' + P)

# Wins to losses
def winsnloss(dataframe, WL):
    return print(dataframe[WL].value_counts())

```
